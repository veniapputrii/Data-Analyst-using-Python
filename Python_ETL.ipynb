```python
#15. Feb
import pandas as pd
#Tabelle 1. Die Mitarbeiter
daten_m = {
    'Name' : ['Anna', 'Ben', 'Clara', 'David'],
    'Team_Code' : [1,2,1,3] #Das ist der Schlüssel
}

df_mitarbeitr = pd.DataFrame(daten_m)

#Tabelle 2 : Die Team-Namen
daten_t = {
    'ID' : [1,2,3],
    'Team_Name' : ['IT', 'Sales', 'HR'] #Die Übersetzung für den Code
}

df_teams = pd.DataFrame(daten_t)

print("___Tabelle 1: Mitarbeiter ___")
display(df_mitarbeitr)
print("___Tabelle 2: Teams ___")
display(df_teams)
```

```python
# 16.Feb
df_komplett = pd.merge(
    df_mitarbeitr,
    df_teams,
    left_on = 'Team_Code',
    right_on= 'ID'
)

display(df_komplett)
```

```python
#17. Feb
daten_m = {
    'Name' : ['Anna','Ben','Clara','David', 'Emil'],
    'Team_Code' : [1,2,1,3,99],
    'Stadt' : ['Berlin', 'Hamburg', 'München', 'Berlin', 'Frankfurt'],
    'ID'  : [1,2,3,4,5]
}
df_mitarbeiter_neu = pd.DataFrame(daten_m)
print(df_mitarbeiter_neu)
```

```python
#21. Feb. 2026
#ICh suche : Team_Code ist 1 UND der Name ist 'Anna'

filter_kombination = (df_mitarbeiter_neu['Team_Code'] == 1) & (df_mitarbeiter_neu['Name'] == 'Anna')

ergebnis = df_mitarbeiter_neu[filter_kombination]
print(ergebnis)
```

```python
#Spalte entfernen
#axis = 0 -> Zeile löschen
#axis = 1 -> Spalte löschen

df_mitarbeiter_neu = id.drop(['Team_Name', 'Team_Code_y', 'Name_y', 'ID_x'], axis=1)
print(df_mitarbeiter_neu)
```

```python
#22. Feb. 2026
import pandas as pd
#1. Unsere bisherige Tabelle (sauberer Neustart)
daten_m = {
    'Name' : ['Anna', 'Ben', 'Clara', 'David', 'Emil'],
    'Team_Code' : [1,2,1,3,4],
    'Abteilung' : ['IT', 'Sales', 'IT', 'HR', 'Sales']
}
df_mitararbeiter = pd.DataFrame(daten_m)

#2. Ein neues Dictionary mitt den euen Mitarbeitern erstellen

neue_daten = {
    'Name' : ['Fion', 'Gustav', 'Hanna', 'Ida', 'Jan'],
    'Team_Code' : [1,2,1,3,4],
    'Abteilung' : ['IT', 'Sales', 'Marketing', 'IT', 'Sales']
}
df_neu = pd.DataFrame(neue_daten)

#3. Beide Tabellen zusammenfügen
#Die eckigen Klammern [ ] sin wichtig! ignore_index=True sorgt dafür,
#dass die zeilennummern (0,1,2,...) sauber weiterzählen

df_komplett = pd. concat([df_mitararbeiter, df_neu], ignore_index=True)

print(df_komplett)
```

```python
#23. Feb. 2026
#Ich filtere wieder nur die Leute aus Team 1 heraus
team_1 = df_komplett[df_komplett['Team_Code'] == 1]

#2. Ich lasse Python zählen, wie oft jede Abteilung in Team 1 vorkommt
verteilung_team_1 = team_1['Abteilung'].value_counts()

print('Verteilung in Team is :')
print(verteilung_team_1)
```

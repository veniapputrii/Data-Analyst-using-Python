```python
#heute lerne ich am 28.Jan
#impotiere das Werkzeug 'Pandas'
import pandas as pd

#erstelle einfache Daten (ein Wörterbuch)
daten = {
    'Name' : ['Anna', 'Bob', 'Charlie', 'David', 'Eva'],
    'Alter': [24, 30, 22, 35, 30],
    'Stadt': ['Berlin', 'München', 'Hamburg', 'Köln', 'Frankfurt'],
    'Gehalt': [6000, 4200, 2800, 3400, 1300]
}

#handle die Daten in eine Tabelle um (einen 'DateFrame')
df = pd.DataFrame(daten)

#Zeige die ersten Zeilen an
print('hier ist deine Tabelle:')
display(df)

#eine erste kleine Analyse: wie hoch ist das Durchschnittsgehalt?
durchschnitt = df['Gehalt'].mean()
print(f'\nDas Durchschnittliche Gehalt beträgt: {durchschnitt} Euro')
```

```python
#Stell dir vor, du wurdest gerade als Datenanalyst von einer Fitnessstudio-Kette eingestellt.
#An deinem ersten Tag gibt dir der Chef eine Liste von  Mitgliedern.
#Das Problem : Die Liste wurde von Hand abgtippt
#und ist voller Lücken und absurder Tippfehler!

#Schritt 1 :
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Unser chaotischer Datensatz
daten_gym = {
    'Mitglied_Name' : ['Anna', 'Ben', 'Clara', 'David', 'Emil', 'Fiona', 'Gustav', 'Hanna'],
    'Alter' : [25, 34, np.nan, 45, 999, 22, 31, np.nan], #np.nan bedeutet : Wert fehlt! 999 ist ein Tippfehler
    'Tarif' : ['Basic', 'Premium', 'Premium', 'Basic', 'Premium', 'Basic', 'Premium','Basic'],
    'Besuche_Pro_Monat' : [4,12,8,15,np.nan,10,5,2]
}
df_gym = pd.DataFrame(daten_gym)
print('Das ist unsere chaoticshe Ausganfstabelle :')
print(df_gym)

#Aufgabe 1 : Fehler korrigieren : Überschreibe die 999 mit ein 29
df_gym['Alter'] = df_gym['Alter'].replace(999, 29)

print(daten_gym)

#Aufgabe 2: Lücken löschen
df_gym = df_gym.dropna(subset=['Alter', 'Besuche_Pro_Monat'])

print(df_gym)
#Aufgaben 3 : Daten analysieren
df_gym = df_gym.groupby('Tarif').mean('Besuche_Pro_Monat')
print(df_gym)


#Visualisieren & Speichern
df_gym.plot(kind='bar', x='Alter', y='Besuche_Pro_Monat')
plt.xlabel('Alter')
plt.ylabel('Durchschnittliche Besuche pro Monat')
plt.show(df_gym)
```
